```python
import requests
from bs4 import BeautifulSoup

def fetch_and_parse_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all rows in the table
    rows = soup.find_all("tr")[1:]  # Skip the header row
    data = []

    #extract x-coordinate, Character, and y-coordinate
    for row in rows:
        cols = row.find_all("td")
        if len(cols) == 3:  # Ensure it has 3 columns
            x = int(cols[0].text.strip())
            char = cols[1].text.strip()
            y = int(cols[2].text.strip())
            data.append({"x-coordinate": x, "y-coordinate": y, "Character": char})
    return data

def decode_secret_message(data):
    # getting grid size
    max_x = max(point['x-coordinate'] for point in data)
    max_y = max(point['y-coordinate'] for point in data)

    #empty grid with spaces
    grid = [[' ' for _ in range(max_y + 1)] for _ in range(max_x + 1)]

    #Populating the grid with characters
    for point in data:
        x = point['x-coordinate']
        y = point['y-coordinate']
        grid[x][y] = point['Character']

    #printing the grid row by row
    for row in grid:
        print("".join(row))
```
